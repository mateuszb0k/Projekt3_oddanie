Filtracja 1d

```python
#test filtracji 1d z przykladowymi kernelami
import cmake_example as ce
inp = [1, 2, 3, 4, 5]
kernel=[1] #identity kernel
expected=[1,2,3,4,5]
out=ce.filter1D(inp, kernel)
print(out)
assert out == expected
kernel =[0.5,0.5] #smoothing kernel
expected=[1.5,2.5,3.5,4.5,2.5]
out=ce.filter1D(inp, kernel)
print(out)
assert out == expected
kernel=[1,-1] #difference kernel
expected=[-1,-1,-1,-1,5]
out=ce.filter1D(inp, kernel)
print(out)
assert out == expected
```

filtracja 2d
do filtracji 2d zaimplementowano 4 filtry
edge detection
gaussian blur
box blur
sharpen

```python
from PIL import Image
import numpy as np
import cmake_example as ce
selection =1 #input nie dzialal w jupyterze
img = Image.open('Vd-Orig.png')
#output = Image.open('Vd-Orig.png')
#output.show()
img_np = np.array(img)
if selection == 1:
	img_blurred = ce.filter2D(img_np, ce.gaussianBlurKernel())
	img_blurred_np = np.array(img_blurred)
	img_blurred_scaled = (img_blurred_np - img_blurred_np.min()) / (img_blurred_np.max() - img_blurred_np.min())
	output= Image.fromarray((img_blurred_scaled*255).astype(np.uint8))
	output.show()
elif selection == 2:
	img_box_blurred = ce.filter2D(img_np, ce.boxblurKernel())
	img_box_blurred_np = np.array(img_box_blurred)
	img_box_blurred_scaled = (img_box_blurred_np - img_box_blurred_np.min()) / (img_box_blurred_np.max() - img_box_blurred_np.min())
	output= Image.fromarray((img_box_blurred_scaled*255).astype(np.uint8))
	output.show()
elif selection == 3:
	img_sharpened = ce.filter2D(img_np, ce.sharpenKernel())
	img_sharpened_np = np.array(img_sharpened)
	img_sharpened_scaled = (img_sharpened_np - img_sharpened_np.min()) / (img_sharpened_np.max() - img_sharpened_np.min())
	output= Image.fromarray((img_sharpened_scaled*255).astype(np.uint8))
	output.show()
#test filtracji 2d na przykladowym zdjeciu 
```

EDGE DETECTION

```python
import numpy as np
from PIL import Image
import cmake_example

img = Image.open('Vd-Orig.png')
img_np = np.array(img)

img_blurred = cmake_example.filter2D(img_np, cmake_example.gaussianBlurKernel())
img_blurred_np = np.array(img_blurred)
img_blurred_scaled = (img_blurred_np - img_blurred_np.min()) / (img_blurred_np.max() - img_blurred_np.min())

img_filtered = cmake_example.filter2D(img_blurred_scaled, cmake_example.edgeDetectionKernel())
img_filtered_np = np.array(img_filtered)
img_filtered_scaled = (img_filtered_np - img_filtered_np.min()) / (img_filtered_np.max() - img_filtered_np.min())

filtered_scaled_uint8 = (img_filtered_scaled * 255).astype(np.uint8)
output = Image.fromarray(filtered_scaled_uint8)
output.show()
```
